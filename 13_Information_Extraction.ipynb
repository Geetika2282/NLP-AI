```python
from urllib.request import urlopen
```

```python
url = urlopen("https://en.wikipedia.org/wiki/Matheran")
```

```python
html = url.read() # url.read - downloads the complete source code (html) of url
```

```python
html
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser') #returns html file
soup
```

```python
soup.find('title') # returns the title tag
```

```python
data = soup.get_text() # removes html tags and only gives the text; strip=True, removes white spaces
```

```python
print(data)
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(data)
tokens
```

```python
from nltk import FreqDist
freq = FreqDist(tokens)
freq.plot(20)
```

```python
from nltk.corpus import stopwords
tokens1 = [token for token in tokens if token.isalpha() or token.isdigit()]
```

```python
tokens1
```

```python
freq = FreqDist(tokens1)
freq.plot(20)
```

```python
swords = stopwords.words('english')
tokens2 = [word.lower() for word in tokens1 if word.lower() not in swords]
```

```python
freq = FreqDist(tokens2)
freq.plot(20);
```

## Tiger - wiki


```python
# Information retreival: load data using web scraping
from nltk.stem import PorterStemmer
ls = PorterStemmer()

url = urlopen('https://en.wikipedia.org/wiki/Tiger')
data = url.read()
soup = BeautifulSoup(data,'html.parser')
text = soup.get_text()
tokens = word_tokenize(text)
tokens1 = [token for token in tokens if token.lower() not in swords]
tokens2 = [token.lower() for token in tokens1 if token.isalnum()]

tokens3 = [ls.stem(token) for token in tokens2 if len(token) > 1]
freq = FreqDist(tokens3)
freq.plot(20)
freq.most_common(5)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

# Step 1: Scrape the webpage
url_noam = urlopen('https://en.wikipedia.org/wiki/Noam_Chomsky')
data = url_noam.read()
soup = BeautifulSoup(data, 'html.parser')

# Step 2: Extract and tokenize text
text = soup.get_text()
tokens = word_tokenize(text)

# Step 3: Remove stopwords
swords = set(stopwords.words('english'))
tokens1 = [token for token in tokens if token.lower() not in swords]

# Step 4: Remove punctuation and perform lemmatization
ls = WordNetLemmatizer()
tokens2 = [token.lower() for token in tokens1 if token.isalnum()]
tokens3 = [ls.lemmatize(token) for token in tokens2 if len(token) > 1]

# Step 5: Frequency Distribution
freq = FreqDist(tokens3)
freq.plot(20, title="Top 20 Frequent Words")
print("Top 5 Most Common Words:", freq.most_common(5))

freq = FreqDist(tokens3)
freq.plot(20)
freq.most_common(5)
```
