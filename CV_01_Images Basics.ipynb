#### Color Channels

```python
plt.figure(figsize=(9,3))
plt.subplot(1,3,1)
plt.title('cv2')
plt.imshow(img)

plt.subplot(1,3,2)
plt.title('matplotlib')
plt.imshow(img_plt)

plt.subplot(1,3,3)
plt.title('scikit-image')
plt.imshow(img_ski)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(9,3))
plt.subplot(1,3,1)
plt.title('cv2')
plt.imshow(img_rgb)

plt.subplot(1,3,2)
plt.title('matplotlib')
plt.imshow(img_plt)

plt.subplot(1,3,3)
plt.title('scikit-image')
plt.imshow(img_ski)
```
