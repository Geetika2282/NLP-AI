```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/circle.png',0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
help(cv2.cornerHarris)
```

```python
corners = cv2.cornerHarris(img, blockSize=2, ksize=3, k=0.04)
```

```python
corners
```

```python
plt.imshow(corners,cmap='gray')
```

```python
new = corners < corners.max()/2
```

```python
plt.imshow(new,cmap='gray')
```

```python
img1 = cv2.imread('images/sachin.jpg',0)
corners1 = cv2.cornerHarris(img1,blockSize=2, ksize=3, k=0.04)

new =  corners1 < corners1.max()/2
plt.imshow(new,cmap='gray')
```
