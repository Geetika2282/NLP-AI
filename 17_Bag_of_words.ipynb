#### Creating a dataframe of count of words

```python
from nltk.tokenize import  word_tokenize
import pandas as pd
import nltk
nltk.download('punkt_tab')

sent1 = "It is a good practice for us"
sent2 = "It was also good to know about it"
sent1 = sent1.lower()
sent2 = sent2.lower()
words= word_tokenize(sent1) + word_tokenize(sent2)
lst = []
for i in words:
  lst.append(i)
words = lst
words = pd.Series(words).unique()
words
print(words)

bag = pd.DataFrame({'words' : words})
bag



s1 = []
s2= []
tokenized_sent1 = word_tokenize(sent1)
tokenized_sent2 = word_tokenize(sent2)

for word in words:
    s1.append(tokenized_sent1.count(word))  # Count in tokenized list
    s2.append(tokenized_sent2.count(word))  # Count in tokenized list

bag['sent1'] = s1
bag['sent2'] = s2




bag.set_index('words',inplace=True)
bag

bag_transposed = bag.transpose()
bag_transposed
```

```python
#Create a set of words for both the sentences
tokens = word_tokenize(sent1.lower()) + word_tokenize(sent2.lower())
tokens
```

```python
#create tokens set
tokens = set(tokens)
tokens
```

```python
#create dataframes with tokens as column names
df = pd.DataFrame({}, index=[1,2], columns= list(tokens))
df
```

```python
#extract tokens from respective sentences
tokens1 = word_tokenize(sent1.lower())
tokens2 = word_tokenize(sent2.lower())
```

```python
tokens1
```

```python
tokens2
```

```python
#extract the counts of words for sentences
counts1 = [tokens1.count(x) for x in df.columns]
counts2 = [tokens2.count(x) for x in df.columns]
```

```python
#supply the counts to the rows of DataFrame
df.iloc[0] = counts1
df.iloc[1] = counts2
```

### Using in-built library

```python
cvt = CountVectorizer()
```

```python
x_new = cvt.fit_transform([sent1,sent2])
```
