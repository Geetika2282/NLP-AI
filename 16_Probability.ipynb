## Therotical Probability

```python
def coin_flip_prob():
  heads_prob = 1/2
  return heads_prob

def rolling_dic_prob(num):
  if num not in range(1, 7):
    raise ValueError("Number must be between 1 and 6")
  prob = 1/6
  return prob
```

```python
coin_flip_prob()
```

```python
# Example usage
coin_flip_res = coin_flip_prob()
dice_roll_res = rolling_dic_prob(2)

print("Probability of heads in a coin flip: ", (coin_flip_res))
print("Probability of rolling a 2 on a dice: ", (dice_roll_res))
```

## Experimental probability

```python
import numpy as np
import random
```

```python
def simulate_coin_flip(num_flips):
  results = []
  for _ in range(num_flips):
    result = random.choice(["heads", "tails"])
    results.append(result)
  return results
```

```python
simulate_coin_flip(10)
```

```python
import pandas as pd
```

```python
pd.Series(simulate_coin_flip(10)).value_counts()
```

```python
from collections import Counter
Counter(simulate_coin_flip(1000))
```

```python
def calculate_experimental_probability(event,results):
  event_count = results.count(event)
  total_flips = len(results)
  probability = event_count / total_flips
  return probability
```

```python
# Example usage

num_flips = 1000
simulation_results = simulate_coin_flip(num_flips)

# Pass 'heads' and 'tails' as the first argument to calculate_experimental_probability
heads_probability = calculate_experimental_probability('heads', simulation_results)
tails_probability = calculate_experimental_probability('tails', simulation_results)

print(f'Number of flips: {num_flips}')
print(f'Heads probability: {heads_probability:.4f}')
print(f'Tails probability: {tails_probability:.4f}')
```
